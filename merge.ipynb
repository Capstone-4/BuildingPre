건축물대장 PK 조인관련입니다.

1. 건축물대장 총괄표제부
서울시 건축물대장 정보.관리_건축물대장_PK = 서울시 건축물대장 총괄표제부 정보.관리_건축물대장_PK

2. 건축물대장 표제부
서울시 건축물대장 정보.관리_건축물대장_PK = 서울시 건축물대장 표제부 정보.관리_건축물대장_PK

3. 건축물대장 전유부
서울시 건축물대장 정보.관리_건축물대장_PK = 서울시 건축물대장 전유부 정보.관리_건축물대장_PK

4. 건축물대장 전유공유면적표
서울시 건축물대장 정보.관리_건축물대장_PK = 서울시 건축물대장 전유공유면적표.관리_건축물대장_PK

※ 참고로 서울시 건축물대장 관리_상위_건축물대장_PK는 표제부의 내용(대지면적, 건축면적, 용적율, 층수, 높이등)을 조인할때 사용함
-표제부와 총괄표제부 또는 전유부을 조인할때 사용

```python
import pandas as pd
```

```python
building = pd.read_csv("pre/건축물대장.csv")
pyo = pd.read_csv("pre/표제부.csv")
chongpyo = pd.read_csv("pre/총괄표제부.csv")
jeon = pd.read_csv("pre/전유부.csv")
floor = pd.read_csv("pre/층별개요.csv")
jeonArea = pd.read_csv("pre/전유공용면적.csv")
```

```python
print(building.columns)
print(pyo.columns)
print(chongpyo.columns)
print(jeon.columns)
print(floor.columns)
print(jeonArea.columns)
```

```python
# Rename Columns
pyo.rename(columns = {"건축물대장 관리번호": "관리_건축물대장_PK", "승강기 수": "승강기_수", "주차 대수": "주차_대수"}, inplace = True)
chongpyo.rename(columns = {"건축물대장 관립번호": "관리_건축물대장_PK", "주차 대수": "주차_대수"}, inplace = True)
jeonArea.rename(columns = {"관리_전유_공용_면적_pk": "관리_건축물대장_PK"}, inplace = True)
```

```python
print(building.columns)
print(pyo.columns)
print(chongpyo.columns)
print(jeon.columns)
print(floor.columns)
print(jeonArea.columns)
```

<h1>Inner Join - 표제부, 총괄표제부, 전유부 중심 3개의 table</h1>

```python
full_pyo = pd.merge(building, pyo, on = "관리_건축물대장_PK", how = "inner")
full_chongpyo = pd.merge(building, chongpyo, on="관리_건축물대장_PK", how="inner")
full_jeon = pd.merge(building, jeon, on="관리_건축물대장_PK", how="inner")
```

```python
print(f"full pyo: {full_pyo.shape}\nfull_chong: {full_chongpyo.shape}\nfull_jeon: {full_jeon.shape}\n")
```

```python
full_pyo.shape[0] + full_chongpyo.shape[0] + full_jeon.shape[0] < building.shape[0]
```

```python
pd.merge(full_chongpyo, full_jeon, left_on="관리_상위_건축물대장_PK", right_on="관리_건축물대장_PK", how="inner")
```

<h1>Outer Join - 건축물 대장 중심으로 모조리 left join한 1 table</h1>

```python
outer = pd.merge(building, pyo, on = "관리_건축물대장_PK", how = "left")
outer.head()
```

```python
outer = pd.merge(outer, chongpyo, on="관리_건축물대장_PK", how="left")
outer.head()
```

```python
outer_jeon = pd.merge(jeon, jeonArea, on="관리_건축물대장_PK", how="inner")
outer_jeon.head(1)
```

```python
outer = pd.merge(outer, outer_jeon, on="관리_건축물대장_PK", how="left")
outer.head()
```

```python
outer.columns
```

```python
outer.shape
```

```python
outer = outer.dropna(subset=['용적율_x', '지상층수', '지하층수', '승강기_수', '주차_대수_x',
                       '용적율_y', '주건축물수', '세대수', '가구수', '주차_대수_y',
                       '동명칭', '호_명', '층_구분_코드_x', '층_구분_코드_y', '층_번호', '면적'], how='all')
outer.head()
```

```python
outer['용적율'] = outer[['용적율_x', '용적율_y']].fillna(0).sum(axis=1)
outer.drop(["용적율_x", "용적율_y"], axis=1, inplace=True)
outer.head()
```

```python
outer['주차_대수'] = outer[['주차_대수_x', '주차_대수_y']].fillna(0).sum(axis=1)
outer.drop(["주차_대수_x", "주차_대수_y"], axis=1, inplace=True)
outer.head()
```
